```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np

# Dummy data (ganti nanti dengan real)
data = {
    'Tahun': [2022, 2023, 2024, 2025],
    'Sales': [4500000000, 4700000000, 5000000000, 4800000000]
}
df = pd.DataFrame(data)

# Linear Regression
X = df[['Tahun']]
y = df['Sales']
model = LinearRegression()
model.fit(X, y)

# Prediksi 2022-2026
tahun_pred = pd.DataFrame({'Tahun': np.arange(2022, 2027)})
tahun_pred['Prediksi'] = model.predict(tahun_pred[['Tahun']])

# Plot
plt.figure(figsize=(8,5))
sns.lineplot(x='Tahun', y='Sales', data=df, marker='o', label='Actual')
sns.lineplot(x='Tahun', y='Prediksi', data=tahun_pred, marker='x', linestyle='--', color='red', label='Linear Regression')

plt.title('Sales Forecast 2022-2026 (Linear Regression)')
plt.ylabel('Sales (Rp)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np

# === DATASET 1: Realistic (campur naik turun) ===
data1 = {
    'Tahun': [2022, 2023, 2024, 2025],
    'Sales': [4500000000, 4700000000, 5000000000, 4800000000]
}
df1 = pd.DataFrame(data1)

# === DATASET 2: Trend Naik ===
data2 = {
    'Tahun': [2022, 2023, 2024, 2025],
    'Sales': [4000000000, 4500000000, 5000000000, 5500000000]
}
df2 = pd.DataFrame(data2)

# === DATASET 3: Trend Turun ===
data3 = {
    'Tahun': [2022, 2023, 2024, 2025],
    'Sales': [5500000000, 5000000000, 4500000000, 4000000000]
}
df3 = pd.DataFrame(data3)

# === Function untuk plot ===
def plot_forecast(df, title):
    X = df[['Tahun']]
    y = df['Sales']
    model = LinearRegression()
    model.fit(X, y)

    tahun_pred = pd.DataFrame({'Tahun': np.arange(2022, 2027)})
    tahun_pred['Prediksi'] = model.predict(tahun_pred[['Tahun']])

    plt.figure(figsize=(8,5))
    sns.lineplot(x='Tahun', y='Sales', data=df, marker='o', label='Actual')
    sns.lineplot(x='Tahun', y='Prediksi', data=tahun_pred, marker='x', linestyle='--', color='red', label='Regression')

    plt.title(title)
    plt.ylabel('Sales (Rp)')
    plt.grid(True)
    plt.legend()
    plt.show()

# === Plot semua ===
plot_forecast(df1, 'Sales Forecast (Realistic Pattern)')
plot_forecast(df2, 'Sales Forecast (Uptrend)')
plot_forecast(df3, 'Sales Forecast (Downtrend)')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np

# Dataset contoh
data = {
    'Tahun': [2022, 2023, 2024, 2025],
    'Sales': [4500000000, 4700000000, 5000000000, 4800000000]
}
df = pd.DataFrame(data)

# Linear regression
X = df[['Tahun']]
y = df['Sales']
model = LinearRegression()
model.fit(X, y)

# Prediksi
tahun_pred = pd.DataFrame({'Tahun': np.arange(2022, 2027)})
tahun_pred['Prediksi'] = model.predict(tahun_pred[['Tahun']])

# Rata-rata
avg_sales = df['Sales'].mean()

# Plot
plt.figure(figsize=(8,5))
sns.lineplot(x='Tahun', y='Sales', data=df, marker='o', label='Actual')
sns.lineplot(x='Tahun', y='Prediksi', data=tahun_pred, marker='x', linestyle='--', color='red', label='Regression')
plt.axhline(avg_sales, color='green', linestyle=':', label=f'Average: Rp {avg_sales:,.0f}')

plt.title('Sales Forecast with Average Line')
plt.ylabel('Sales (Rp)')
plt.grid(True)
plt.legend()
plt.show()
```

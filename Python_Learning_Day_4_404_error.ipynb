```python
# Langkah 1: Install library yang diperlukan
!pip install tensorflow matplotlib

# Langkah 2: Import library
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np

# Langkah 3: Load dataset MNIST
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalisasi data agar nilai piksel berada di rentang [0, 1]
x_train, x_test = x_train / 255.0, x_test / 255.0

# Langkah 4: Membuat model Neural Network sederhana
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),  # Meluruskan gambar 28x28 menjadi vektor 1D
    layers.Dense(128, activation='relu'),  # Layer tersembunyi dengan 128 neuron dan aktivasi ReLU
    layers.Dropout(0.2),                   # Dropout untuk mengurangi overfitting
    layers.Dense(10, activation='softmax') # Output layer dengan 10 kelas (angka 0-9)
])

# Langkah 5: Compile model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Langkah 6: Visualisasi arsitektur model
model.summary()

# Langkah 7: Latih model
history = model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))

# Langkah 8: Evaluasi model
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f"\nAkurasi pada data uji: {test_acc:.2f}")

# Langkah 9: Visualisasi hasil pelatihan
plt.figure(figsize=(12, 4))

# Plot akurasi
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Akurasi Pelatihan')
plt.plot(history.history['val_accuracy'], label='Akurasi Validasi')
plt.title('Akurasi Model')
plt.xlabel('Epoch')
plt.ylabel('Akurasi')
plt.legend()

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Loss Pelatihan')
plt.plot(history.history['val_loss'], label='Loss Validasi')
plt.title('Loss Model')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.show()

# Langkah 10: Visualisasi prediksi model
predictions = model.predict(x_test)

def plot_image(i, predictions_array, true_label, img):
    """Fungsi untuk menampilkan gambar dan prediksi."""
    predictions_array, true_label, img = predictions_array[i], true_label[i], img[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])

    plt.imshow(img, cmap=plt.cm.binary)

    predicted_label = np.argmax(predictions_array)
    if predicted_label == true_label:
        color = 'blue'
    else:
        color = 'red'

    plt.xlabel(f"Pred: {predicted_label} (True: {true_label})", color=color)

def plot_value_array(i, predictions_array, true_label):
    """Fungsi untuk menampilkan diagram batang probabilitas prediksi."""
    predictions_array, true_label = predictions_array[i], true_label[i]
    plt.grid(False)
    plt.xticks(range(10))
    plt.yticks([])
    thisplot = plt.bar(range(10), predictions_array, color="#777777")
    plt.ylim([0, 1])
    predicted_label = np.argmax(predictions_array)

    thisplot[predicted_label].set_color('red')
    thisplot[true_label].set_color('blue')

# Menampilkan beberapa hasil prediksi
num_rows = 5
num_cols = 3
num_images = num_rows * num_cols
plt.figure(figsize=(2 * 2 * num_cols, 2 * num_rows))
for i in range(num_images):
    plt.subplot(num_rows, 2 * num_cols, 2 * i + 1)
    plot_image(i, predictions, y_test, x_test)
    plt.subplot(num_rows, 2 * num_cols, 2 * i + 2)
    plot_value_array(i, predictions, y_test)
plt.tight_layout()
plt.show()
```

```python
# Install library yang diperlukan
!pip install tensorflow matplotlib

# Import library
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation

# Load dataset MNIST
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalisasi data
x_train, x_test = x_train / 255.0, x_test / 255.0

# Membuat model Neural Network sederhana
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(10, activation='softmax')
])

# Compile model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Fungsi untuk melatih model dan menyimpan sejarah pelatihan
class TrainingHistory:
    def __init__(self):
        self.accuracy = []
        self.loss = []

    def on_epoch_end(self, epoch, logs):
        self.accuracy.append(logs['accuracy'])
        self.loss.append(logs['loss'])

# Inisialisasi objek untuk menyimpan sejarah pelatihan
history = TrainingHistory()

# Latih model dengan callback untuk menyimpan sejarah pelatihan
model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test),
          callbacks=[tf.keras.callbacks.LambdaCallback(on_epoch_end=history.on_epoch_end)])

# Visualisasi animasi
fig, ax = plt.subplots(figsize=(8, 4))
ax.set_xlim(0, 10)
ax.set_ylim(0, 1)
line_accuracy, = ax.plot([], [], lw=2, label='Akurasi Pelatihan')
line_loss, = ax.plot([], [], lw=2, label='Loss Pelatihan')
ax.legend()

# Fungsi inisialisasi untuk animasi
def init():
    line_accuracy.set_data([], [])
    line_loss.set_data([], [])
    return line_accuracy, line_loss

# Fungsi update untuk animasi
def update(frame):
    x = list(range(1, frame + 2))  # Epoch ke-n
    y_accuracy = history.accuracy[:frame + 1]
    y_loss = history.loss[:frame + 1]
    line_accuracy.set_data(x, y_accuracy)
    line_loss.set_data(x, y_loss)
    return line_accuracy, line_loss

# Buat animasi
ani = FuncAnimation(fig, update, frames=10, init_func=init, blit=True)

# Simpan animasi sebagai GIF
ani.save('neural_network_training.gif', writer='imagemagick', fps=2)

# Tampilkan animasi di notebook
plt.show()
```

```python
# Install library yang diperlukan
!pip install tensorflow matplotlib numpy

# Import library
import tensorflow as tf
from tensorflow.keras.layers import MultiHeadAttention, Dense, LayerNormalization
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation

# Simulasi input data
sequence_length = 10  # Panjang urutan (misalnya kalimat dengan 10 kata)
embedding_dim = 8     # Dimensi embedding (fitur per kata)

# Buat input dummy
x = tf.random.uniform((1, sequence_length, embedding_dim))  # Batch size 1

# Membuat layer MultiHeadAttention
multi_head_attention = MultiHeadAttention(num_heads=2, key_dim=embedding_dim)

# Forward pass untuk mendapatkan attention weights
output, attention_weights = multi_head_attention(x, x, return_attention_scores=True)

# Ekstrak attention weights
attention_weights = attention_weights.numpy().squeeze()  # Shape: (num_heads, seq_len, seq_len)

# Visualisasi animasi attention weights
fig, ax = plt.subplots(figsize=(6, 6))
im = ax.imshow(attention_weights[0], cmap='viridis', interpolation='nearest')
ax.set_title("Attention Weights (Head 1)")
cbar = fig.colorbar(im)

# Fungsi inisialisasi untuk animasi
def init():
    im.set_data(np.zeros_like(attention_weights[0]))
    return [im]

# Fungsi update untuk animasi
def update(frame):
    im.set_data(attention_weights[frame % len(attention_weights)])  # Loop melalui setiap head
    ax.set_title(f"Attention Weights (Head {frame % len(attention_weights) + 1})")
    return [im]

# Buat animasi
ani = FuncAnimation(fig, update, frames=len(attention_weights) * 3, init_func=init, blit=True, interval=1000)

# Simpan animasi sebagai GIF
ani.save('transformer_attention.gif', writer='imagemagick', fps=1)

# Tampilkan animasi di notebook
plt.show()
```

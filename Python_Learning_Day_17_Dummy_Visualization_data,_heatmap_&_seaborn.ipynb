```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Set style untuk Matplotlib
plt.style.use('seaborn-v0_8-notebook')

# Membuat dataset penjualan sederhana
data = {
    'bulan': ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun'],
    'produk_A': [150, 200, 180, 220, 250, 300],
    'produk_B': [100, 120, 130, 150, 180, 200],
    'produk_C': [80, 90, 110, 130, 150, 170]
}

df = pd.DataFrame(data)
df

# Membuat grafik garis untuk tren penjualan
plt.figure(figsize=(10, 6))
plt.plot(df['bulan'], df['produk_A'], marker='o', label='Produk A')
plt.plot(df['bulan'], df['produk_B'], marker='s', label='Produk B')
plt.plot(df['bulan'], df['produk_C'], marker='^', label='Produk C')

plt.title('Tren Penjualan Produk A, B, dan C')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penjualan')
plt.legend()
plt.grid(True)
plt.show()

# Membuat grafik batang untuk perbandingan penjualan per produk
plt.figure(figsize=(10, 6))
width = 0.2
x = np.arange(len(df['bulan']))

plt.bar(x - width, df['produk_A'], width, label='Produk A')
plt.bar(x, df['produk_B'], width, label='Produk B')
plt.bar(x + width, df['produk_C'], width, label='Produk C')

plt.title('Perbandingan Penjualan Produk per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penjualan')
plt.xticks(x, df['bulan'])
plt.legend()
plt.grid(axis='y')
plt.show()

# Membuat pie chart untuk persentase penjualan total per produk
total_sales = df[['produk_A', 'produk_B', 'produk_C']].sum()
labels = ['Produk A', 'Produk B', 'Produk C']

plt.figure(figsize=(8, 8))
plt.pie(total_sales, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Persentase Penjualan Total per Produk')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()

# Membuat scatter plot untuk analisis hubungan antara penjualan dan diskon
# Anggaplah ada kolom diskon untuk setiap produk
df['diskon_produk_A'] = [10, 15, 5, 20, 10, 5]
df['diskon_produk_B'] = [5, 10, 8, 12, 15, 20]
df['diskon_produk_C'] = [8, 12, 10, 15, 20, 25]

plt.figure(figsize=(10, 6))
plt.scatter(df['diskon_produk_A'], df['produk_A'], color='blue', label='Produk A')
plt.scatter(df['diskon_produk_B'], df['produk_B'], color='green', label='Produk B')
plt.scatter(df['diskon_produk_C'], df['produk_C'], color='red', label='Produk C')

plt.title('Hubungan Antara Diskon dan Penjualan')
plt.xlabel('Diskon (%)')
plt.ylabel('Jumlah Penjualan')
plt.legend()
plt.grid(True)
plt.show()

# Membuat scatter plot untuk analisis hubungan antara penjualan dan diskon
# Anggaplah ada kolom diskon untuk setiap produk
df['diskon_produk_A'] = [10, 15, 5, 20, 10, 5]
df['diskon_produk_B'] = [5, 10, 8, 12, 15, 20]
df['diskon_produk_C'] = [8, 12, 10, 15, 20, 25]

plt.figure(figsize=(10, 6))
plt.scatter(df['diskon_produk_A'], df['produk_A'], color='blue', label='Produk A')
plt.scatter(df['diskon_produk_B'], df['produk_B'], color='green', label='Produk B')
plt.scatter(df['diskon_produk_C'], df['produk_C'], color='red', label='Produk C')

plt.title('Hubungan Antara Diskon dan Penjualan')
plt.xlabel('Diskon (%)')
plt.ylabel('Jumlah Penjualan')
plt.legend()
plt.grid(True)
plt.show()

# Membuat heatmap untuk melihat korelasi antara variabel
correlation_matrix = df[['produk_A', 'produk_B', 'produk_C']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Korelasi Antara Penjualan Produk')
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.colors as mcolors

# Set style untuk Matplotlib
plt.style.use('seaborn-v0_8-whitegrid')

# Membuat dataset survey kepuasan masyarakat
data = {
    'kota': ['Jakarta', 'Bogor', 'Tangerang', 'Depok', 'Bekasi'],
    'kepuasan_sangat': [25, 18, 20, 22, 15],
    'kepuasan_cukup': [35, 40, 30, 38, 45],
    'kepuasan_kurang': [20, 22, 25, 25, 20],
    'kepuasan_tidak_sama_sama_sekali': [20, 20, 25, 15, 20],
    'penduduk_juta': [11, 2.1, 2.3, 2.2, 2.8],
    'ipm': [75.2, 72.1, 74.5, 73.8, 71.5],  # Indeks Pembangunan Manusia
    'kepadatan_penduduk': [15000, 10000, 8000, 9000, 7000]
}

df = pd.DataFrame(data)

# Membuat scatter plot dengan garis regresi, vektor arah, dan palet warna kustom
fig, ax = plt.subplots(figsize=(12, 8))

# Scatter plot dengan hue, size, dan style
scatter = sns.scatterplot(
    x='kepadatan_penduduk',
    y='kepuasan_sangat',
    hue='ipm',
    size='penduduk_juta',
    style='kota',
    data=df,
    palette='viridis',
    sizes=(50, 300),
    alpha=0.8,
    legend='full',
    ax=ax  #  ax 
)

# Menambahkan garis regresi
sns.regplot(
    x='kepadatan_penduduk',
    y='kepuasan_sangat',
    data=df,
    scatter=False,
    color='red',
    line_kws={'lw': 2, 'ls': '--'},
    ax=ax  #  ax 
)

# Menambahkan vektor arah menggunakan quiver
for i, row in df.iterrows():
    ax.arrow(
        row['kepadatan_penduduk'],
        row['kepuasan_sangat'],
        dx=500,
        dy=2,
        width=0.5,
        color='black',
        alpha=0.5,
        length_includes_head=True,
        head_width=1
    )

# Menambahkan anotasi
for i, row in df.iterrows():
    ax.text(
        row['kepadatan_penduduk'],
        row['kepuasan_sangat'],
        row['kota'],
        fontsize=9,
        ha='right'
    )

# Kustomisasi plot
ax.set_title('Kepuasan Masyarakat vs Kepadatan Penduduk dengan IPM dan Populasi', fontsize=14)
ax.set_xlabel('Kepadatan Penduduk (per km)', fontsize=12)
ax.set_ylabel('Persentase Kepuasan Sangat (%)', fontsize=12)
ax.grid(True, linestyle='--', alpha=0.6)
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)

# Menambahkan palet warna kustom untuk hue
norm = plt.Normalize(df['ipm'].min(), df['ipm'].max())
cmap = plt.cm.viridis
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])

# Menambahkan colorbar dengan ax=ax
cbar = fig.colorbar(sm, label='IPM (Indeks Pembangunan Manusia)', pad=0.1, ax=ax)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.colors as mcolors
from matplotlib.animation import FuncAnimation, PillowWriter

# Set style untuk Matplotlib
plt.style.use('seaborn-v0_8-whitegrid')

# Membuat dataset animasi dengan 150 hari kerja
np.random.seed(42)
days = np.arange(1, 151)
kepadatan_penduduk = np.random.randint(7000, 15000, size=150)
kepuasan_sangat = np.random.randint(10, 30, size=150)
ipm = np.random.uniform(70, 80, size=150)
penduduk_juta = np.random.uniform(1, 3, size=150)

df_animasi = pd.DataFrame({
    'hari': days,
    'kepadatan_penduduk': kepadatan_penduduk,
    'kepuasan_sangat': kepuasan_sangat,
    'ipm': ipm,
    'penduduk_juta': penduduk_juta
})

# Membuat animasi scatter plot
fig, ax = plt.subplots(figsize=(12, 8))
scatter = ax.scatter([], [], c=[], s=[], cmap='viridis', alpha=0.8)
title = ax.set_title('Kepuasan Masyarakat Tangsel terhadap Pemerintahan Prabowo-Gibran (Hari 1)', fontsize=14)
ax.set_xlabel('Kepadatan Penduduk (per km)', fontsize=12)
ax.set_ylabel('Persentase Kepuasan Sangat (%)', fontsize=12)
ax.grid(True, linestyle='--', alpha=0.6)

# Menambahkan colorbar
norm = plt.Normalize(df_animasi['ipm'].min(), df_animasi['ipm'].max())
cmap = plt.cm.viridis
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = fig.colorbar(sm, label='IPM (Indeks Pembangunan Manusia)', pad=0.1)

# Fungsi untuk memperbarui animasi
def update(frame):
    # Filter data hingga hari ke-`frame`
    df_frame = df_animasi[df_animasi['hari'] <= frame]

    # Update scatter plot
    scatter.set_offsets(np.c_[df_frame['kepadatan_penduduk'], df_frame['kepuasan_sangat']])
    scatter.set_array(df_frame['ipm'])
    scatter.set_sizes(df_frame['penduduk_juta'] * 100)

    # Update judul
    title.set_text(f'Kepuasan Masyarakat Tangsel terhadap Pemerintahan Prabowo-Gibran (Hari {frame})')

    return scatter, title

# Membuat animasi
ani = FuncAnimation(
    fig=fig,
    func=update,
    frames=150,
    interval=200,  # Kecepatan animasi (ms per frame)
    blit=True
)

# Menyimpan animasi sebagai GIF
writer = PillowWriter(fps=5)  # 5 frame per second
ani.save('kepuasan_masyarakat_tangsel.gif', writer=writer)

plt.close()

!pip install Pillow


```
